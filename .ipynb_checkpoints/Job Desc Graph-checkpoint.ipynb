```python
# re-arrange dataframe 
def expand(row):
    skills = row['job_allSkills'] if isinstance(row['job_allSkills'], list) else [row['job_allSkills']]
    s = pd.Series(row['JobID'], index=list(set(skills)))
    return s

df = apps.apply(expand, axis=1).stack()
df = df.to_frame().reset_index(level=1, drop=False)
df.columns = ['skills', 'JobID']
df.reset_index(drop=True, inplace=True)
```

```python
# create occurrence matrix
occurenceMat = pd.get_dummies(df.skills).groupby(df.JobID).apply(max)
```

```python
occurenceMat.shape
```

```python
# create co-occurrence matrix with Jaccard sim (taken from Boon Teck)
co_occurenceMat = occurenceMat.T.dot(occurenceMat) # occurence dot multiply transpose
np.fill_diagonal(co_occurenceMat.values, 0) # set diagonals co-occurences to zero
# # Normalize each element by row and column sum
# # greater proportion -> smaller distance (closer to 0), so take 1 - (a / (row sum + col sum - a))
# somehow pandas df screws up at some parts of the matrix, converting into np array and calculating there instead
co_occur_np = co_occurenceMat.values
neg_log = np.vectorize(lambda x: -1*log2(x) if ((x > 0) and ~(np.isinf(x))) else 0, otypes = [float]) # lower match, higher weight

# ignore divide by zero warning; correcting it later
np.seterr(divide='ignore', invalid='ignore')
co_occurenceMat = pd.DataFrame(
    neg_log(
        co_occur_np / (
            co_occur_np.sum(axis = 1, keepdims = True) + 
            co_occur_np.sum(axis = 0, keepdims = True) -
            co_occur_np
        )
    ), 
    index = co_occurenceMat.index, 
    columns = co_occurenceMat.index
)

# remove edges of zero weight then create network dataframe
networkDF = co_occurenceMat.replace(to_replace = 0, value = np.nan).stack().reset_index() # values of 0 means no relationship at all
networkDF.rename(columns = {'level_0' : 'source','level_1' : 'target', 0: 'weight'}, inplace = True)
```

```python
# create network graph & find all shortest paths from each skill to every other skill (taken from Boon Teck)
graph_dijkstra = nx.from_pandas_dataframe(networkDF, source = 'source',target = 'target', edge_attr = ['weight'],create_using = nx.Graph())
allShortestPaths = nx.all_pairs_dijkstra_path(graph_dijkstra, weight = 'weight')
shortest_path_lengths = nx.all_pairs_dijkstra_path_length(graph_dijkstra, weight = 'weight')
```

```python
# somehow visualization plot further nodes close to the source, taking 1/weight here to reverse that
networkDF_for_plot = networkDF.copy()
networkDF_for_plot.weight = networkDF_for_plot.weight.apply(lambda x: 1000/x)
graph = nx.from_pandas_dataframe(networkDF_for_plot, source = 'source',target = 'target', edge_attr = ['weight'],create_using = nx.Graph())
```

```python
def nearest_neighbours(skill, label_flag, k):
#     length = [len(allShortestPaths[skill][path]) for path in allShortestPaths[skill]]
#     nearestSkills = [path for path in allShortestPaths[skill] if len(allShortestPaths[skill][path]) == 2]

    skill_dict = shortest_path_lengths[skill].to_dict()
    top_percent = np.percentile(list(skill_dict.values()), 1.5)
    nearestSkills = [k for k in skill_dict if skill_dict[k] < top_percent]
    
    lineWidth = 2
    circleSize = 12
    plot = Plot(plot_width=700, plot_height=550,
                x_range=Range1d(-1.1,1.1), y_range=Range1d(-1.1,1.1),
                tools = [PanTool(), ResetTool(), WheelZoomTool(), SaveTool(), TapTool()]
               )
    plot.title.text = "Figure 4: Nearest Neighbours of skill - " + skill
    
    subG = graph.subgraph([skill] + nearestSkills)
    allEdges = subG.edges()
    [subG.remove_edge(edge[0],edge[1]) for edge in allEdges if edge[0] != skill]
    [subG.add_edge(edge[0],edge[1]) for edge in allEdges if (edge[1] in nx.isolates(subG)) & (edge[0] in allShortestPaths[skill][edge[1]])]
    num_nodes = len(subG.nodes())

    graph_renderer = from_networkx(subG, 
                                   nx.spring_layout,
                                   weight = 'weight', k = k/1000,
#                                    k =  -0.0006 * num_nodes + 0.999, # linear function so that k is lower for fewer nodes
                                   iterations = 70,
                                   scale=1.5, center=(0,0))

    # add color to represent source skill as orange, others as blue
    index = graph_renderer.node_renderer.data_source.data['index']
    graph_renderer.node_renderer.data_source.data['fill_color'] = ['#FF512C'] + np.repeat(Spectral4[0],len(index) - 1).tolist()
    graph_renderer.node_renderer.data_source.column_names += ['fill_color'] # prevents bad column name exception
    graph_renderer.node_renderer.glyph = Circle(size= circleSize, fill_color= 'fill_color')
    
    graph_renderer.node_renderer.hover_glyph = Circle(size= circleSize, fill_color=Spectral4[1])
    
    graph_renderer.edge_renderer.glyph = MultiLine(line_color="#CCCCCC", line_alpha=0.8, line_width= lineWidth)
    
# No need for tap tools
    graph_renderer.node_renderer.selection_glyph = Circle(size= circleSize, fill_color=Spectral4[2])
    graph_renderer.edge_renderer.selection_glyph = MultiLine(line_color=Spectral4[2], line_width= lineWidth)
    graph_renderer.selection_policy = NodesAndLinkedEdges()
    

    if (label_flag is False):
        # add skill label into data source for hover -> columnDatasource saves skill names as index
        plot.add_tools(HoverTool(tooltips=[('Skill','@index')]))
    else:
        # add text labels
        coords = pd.DataFrame([graph_renderer.layout_provider.graph_layout[a] for a in graph_renderer.layout_provider.graph_layout])
        graph_renderer.node_renderer.data_source.data['x'] = coords[0]
        graph_renderer.node_renderer.data_source.data['y'] = coords[1]
        labels = LabelSet(x = 'x',y = 'y',text='index', level='glyph',
                          x_offset=5, y_offset=5, 
                          text_font = '6',
                          source=graph_renderer.node_renderer.data_source, render_mode='canvas')
        plot.add_layout(labels)
    
    plot.renderers.append(graph_renderer)
#     html = file_html(plot, CDN, "Skill Graph: "+ skill)
#     Html_file= open("F:/skillGraph_" + skill + ".html","w")
#     Html_file.write(html)
#     Html_file.close()
    show(plot)

# interact(nearest_neighbours, skill = skill_widget(), label_flag = Checkbox(value = True))

#if need to experiment value of k in from_networkx
#spring_adjust = IntSlider(value=700,min=0,max=1000,step=1) 
#interact(nearest_neighbours, skill = skill_widget(), label_flag = Checkbox(value = True), k = spring_adjust)
```
